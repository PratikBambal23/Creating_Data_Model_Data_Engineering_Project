```python
import pandas as pd
import psycopg2
```

```python
def create_database():
    # Creating connection with Postgres Server
    conn = psycopg2.connect("host=127.0.0.1 dbname=postgres user=postgres password=admin")
    conn.set_session(autocommit=True)
    
    #Creating Cursor to execute SQL queries
    cur = conn.cursor()
    
    # Initialy droping DATABASE accounts if aleady created
    cur.execute("DROP DATABASE IF EXISTS account")    
    cur.execute("CREATE DATABASE account")
    
    # Closing the current connection
    conn.close()
    
    # Connecting to newly created Database Accounts
    conn = psycopg2.connect("host=localhost dbname=account user=postgres password=admin")
    cur = conn.cursor()
    
    return cur, conn
```
